```python
major_corrs = ['ZHVI_AllHomes']
comp = ['ZHVI_1bedroom', 'ZHVI_2bedroom', 'ZHVI_3bedroom', 'ZHVI_4bedroom', 'ZHVI_5BedroomOrMore', 'ZHVI_AllHomes']
corr = data[comp].corr(method='pearson').reindex(major_corrs)
abs(corr.reindex(major_corrs))
```

```python
plt.figure(figsize = (16, 3))
cmap = sns.color_palette("Blues") 
sns.heatmap(corr, 
            xticklabels=corr.columns,
            yticklabels=major_corrs,
            cmap=cmap,
            linewidth=1, linecolor='white')
sns.set(font_scale = 1.5)
plt.show()
```
